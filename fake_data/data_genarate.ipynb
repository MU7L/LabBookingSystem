```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = ['device_name', 'lab_no'])
devices = ['网络', '电子屏幕', '电脑', '摄像头', '音频输入输出设备']
```

```python
# 算法 1-35
# tmpdf = ''
for i in range(1,36):
#     print(i)
    tmpdf = pd.DataFrame({
        'device_name': ['网络', '电子屏幕'],
        'lab_no': [i, i]
    })
    for j in range(20):
        tmpdf.loc[len(tmpdf.index)] = ['电脑', i]
#     break
    df = pd.concat([df, tmpdf], ignore_index = True)
df
```

```python
# 微原 36-70
for i in range(36,71):
    tmpdf = pd.DataFrame({
        'device_name': ['网络', '电子屏幕'],
        'lab_no': [i, i]
    })
    for j in range(20):
        tmpdf.loc[len(tmpdf.index)] = ['电脑', i]
        tmpdf.loc[len(tmpdf.index)] = ['摄像头', i]
        tmpdf.loc[len(tmpdf.index)] = ['音频输入输出设备', i]
    df = pd.concat([df, tmpdf], ignore_index = True)
df
```

```python
# 数据库 71-105
for i in range(71, 106):
    tmpdf = pd.DataFrame({
        'device_name': ['网络', '电子屏幕'],
        'lab_no': [i, i]
    })
    for j in range(30):
        tmpdf.loc[len(tmpdf.index)] = ['电脑', i]
    df = pd.concat([df, tmpdf], ignore_index = True)
df
```
